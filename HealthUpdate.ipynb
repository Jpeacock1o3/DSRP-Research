```python
import pandas as pd
```

```python
url ='https://raw.githubusercontent.com/Jpeacock1o3/DSRP-Research/refs/heads/main/HousingAndCrimeDataset.csv'
BetterHelp = pd.read_csv(url)
```

```python
print("\nActual columns after loading:", BetterHelp.columns.tolist())
```

```python
id_vars = ['REF_AREA', 'Reference area', 'TIME_PERIOD', 'Age', 'Sex', 'Education level', 'Domain', 'Unit of measure']
```

```python
id_vars_present = [col for col in id_vars if col in BetterHelp.columns]
```

```python
Health_map = {'3_3': 'Housing Cost Overburden', '3_2': 'Housing Affordability', '11_1': 'Life Satisfaction', '10_1': 'Homicides', '10_3': 'Road Deaths', '1_5': 'Difficulty making ends meet', '10_2': 'Feeling safe at night', '10_2_DEP': 'Not feeling safe at night'}
```

```python
BetterHelp['OBS_VALUE'] = pd.to_numeric(BetterHelp['OBS_VALUE'], errors='coerce')
BetterHelp.dropna(subset=['OBS_VALUE'], inplace=True)
```

```python
df_wide_aligned = BetterHelp.pivot_table(index=['REF_AREA', 'TIME_PERIOD'],
                                      columns='MEASURE',
                                      values='OBS_VALUE',
                                      aggfunc='mean', # Use mean to aggregate if multiple entries for country/year
                                      fill_value=None # Keep NaN where data is genuinely missing
                                     ).reset_index()
```

```python
df_wide_aligned.columns.name = None
```

```python
columns_to_rename = {code: name for code, name in Health_map.items() if code in df_wide_aligned.columns}
df_wide_aligned.rename(columns=columns_to_rename, inplace=True)
```
